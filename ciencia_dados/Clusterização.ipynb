#### Aprendizado não supervisonado

### Teoria

Como Estudar e Entender o Aprendizado Não Supervisionado
Fundamentos Teóricos:

O que é? Comece entendendo a diferença principal entre aprendizado supervisionado (com rótulos) e não supervisionado (sem rótulos). O objetivo do não supervisionado é descobrir estruturas ocultas nos dados.
Para que serve? Entenda as principais aplicações: clusterização (segmentação de clientes), redução de dimensionalidade (compressão de dados, visualização), detecção de anomalias (fraudes), associação (cestas de compras).
Métricas de Avaliação: Diferente do supervisionado, não temos rótulos para comparar diretamente. Estude métricas intrínsecas como o coeficiente de silhueta, Davies-Bouldin index, ou a soma dos quadrados das distâncias (Elbow Method para K-Means).
Pré-processamento de Dados: A qualidade dos dados é ainda mais crítica no aprendizado não supervisionado. Entenda a importância da normalização, padronização e tratamento de outliers.
Algoritmos Chave:

Clusterização:
K-Means: O mais popular e um excelente ponto de partida. Entenda seu funcionamento iterativo, o papel dos centróides e a sensibilidade à inicialização.
Hierarchical Clustering (Agrupamento Hierárquico): Entenda como ele constrói uma árvore de clusters (dendrograma) e as abordagens aglomerativa e divisiva.
DBSCAN: Importante para lidar com clusters de formatos arbitrários e ruído. Entenda os conceitos de pontos centrais, de fronteira e de ruído.
Gaussian Mixture Models (GMM): Uma abordagem probabilística para clusterização, que assume que os dados são gerados a partir de uma mistura de distribuições gaussianas.
Redução de Dimensionalidade:
PCA (Principal Component Analysis): Fundamental para entender como projetar dados em um espaço de menor dimensão, mantendo a maior variância possível.
t-SNE/UMAP: Para visualização de dados de alta dimensão em 2D ou 3D.
Abordagem Prática:

Linguagem de Programação: Python é a escolha predominante, com bibliotecas como scikit-learn, pandas e matplotlib.
Exercícios Práticos: A melhor forma de aprender é codificando. Implemente os algoritmos do zero (pelo menos as partes mais simples) para entender a lógica.
Conjuntos de Dados de Exemplo: Utilize datasets conhecidos (Iris, Wine, MNIST) para aplicar os algoritmos e visualizar os resultados.
Projetos Reais: Tente aplicar o aprendizado não supervisionado em problemas do mundo real, mesmo que simplificados.
Funcionamento e Exemplos
Vamos focar na clusterização, que é um dos pilares do aprendizado não supervisionado.

1. K-Means

Conceito: O algoritmo K-Means é um dos métodos de clusterização mais populares devido à sua simplicidade e eficiência. Ele tem como objetivo particionar n observações em k clusters, onde cada observação pertence ao cluster cujo centro (médio) é o mais próximo.

Funcionamento:

Inicialização: Escolha aleatoriamente k pontos de dados como os centróides iniciais dos k clusters.
Atribuição: Para cada ponto de dado no dataset, calcule a distância Euclidiana (ou outra métrica de distância) até cada um dos k centróides. Atribua o ponto de dado ao cluster cujo centróide é o mais próximo.
Atualização: Recalcule os centróides de cada cluster. O novo centróide é a média de todos os pontos de dados atualmente atribuídos a esse cluster.
Repetição: Repita os passos 2 e 3 até que os centróides não mudem significativamente entre as iterações, ou um número máximo de iterações seja atingido.

Exemplo (Conceitual):

Imagine que você é um analista de marketing de um e-commerce e tem dados sobre a frequência de compra e o valor gasto por cliente. Você quer segmentar seus clientes para criar campanhas de marketing direcionadas.

Dados: Pontos em um gráfico, onde o eixo X é "Frequência de Compra" e o eixo Y é "Valor Gasto".
Objetivo: Agrupar clientes em k (digamos, 3) segmentos: "Clientes de Alto Valor", "Clientes Regulares", "Clientes Ocasionais".
K-Means em Ação:
O algoritmo escolhe 3 centróides aleatórios.
Cada cliente é atribuído ao centróide mais próximo.
Os centróides se movem para o centro dos grupos de clientes que foram atribuídos a eles.
Os clientes são reatribuídos aos novos centróides mais próximos.
Este processo se repete até que os grupos se estabilizem.
Resultado: Você terá 3 clusters distintos de clientes, permitindo que a equipe de marketing crie estratégias específicas para cada grupo (e.g., promoções de fidelidade para clientes de alto valor, incentivos de recompra para clientes ocasionais).
2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

Conceito: Diferente do K-Means que assume clusters esféricos e de densidade similar, o DBSCAN é um algoritmo de clusterização baseado em densidade. Ele é capaz de encontrar clusters de formas arbitrárias e identificar ruído (outliers) nos dados.

Funcionamento:

Parâmetros: O DBSCAN requer dois parâmetros principais:
epsilon (eps): O raio máximo da vizinhança para buscar outros pontos.
min_samples (MinPts): O número mínimo de pontos dentro do epsilon para que um ponto seja considerado um "ponto central".
Classificação de Pontos:
Ponto Central (Core Point): Um ponto é um ponto central se houver pelo menos min_samples outros pontos (incluindo ele mesmo) dentro de sua vizinhança epsilon.
Ponto de Borda (Border Point): Um ponto é um ponto de borda se estiver dentro da vizinhança epsilon de um ponto central, mas ele próprio não for um ponto central (ou seja, não tem min_samples em sua própria vizinhança).
Ponto de Ruído (Noise Point): Um ponto é um ponto de ruído se não for nem um ponto central nem um ponto de borda (ou seja, não está dentro da vizinhança epsilon de nenhum ponto central).
Formação de Clusters:
O DBSCAN começa com um ponto arbitrário não visitado.
Se for um ponto central, um novo cluster é formado. Todos os pontos densamente alcançáveis (pontos que são pontos centrais ou de borda conectados por uma cadeia de pontos centrais) são adicionados a esse cluster.
Se for um ponto de borda ou de ruído, ele é marcado como visitado e o algoritmo passa para o próximo ponto não visitado.
Este processo continua até que todos os pontos tenham sido visitados.
Exemplo (Conceitual):

Imagine que você está analisando dados de localização de sensores de qualidade do ar em uma cidade. Alguns sensores estão agrupados em bairros densamente povoados, enquanto outros estão espalhados em áreas rurais, e alguns são claramente leituras anômalas (ruído).

Dados: Coordenadas geográficas (latitude, longitude) dos sensores.
Objetivo: Identificar grupos de sensores que representam diferentes áreas geográficas com padrões de qualidade do ar semelhantes, e isolar sensores com leituras atípicas.
DBSCAN em Ação:
Você define epsilon (por exemplo, 500 metros) e min_samples (por exemplo, 3 sensores).
O DBSCAN varre os pontos. Se um sensor tiver pelo menos 3 outros sensores a 500 metros dele, ele se torna um "ponto central".
A partir desses pontos centrais, o algoritmo "expande" os clusters, incluindo outros pontos centrais e de borda que estão dentro de suas vizinhanças.
Sensores isolados, que não estão densamente conectados a outros, são identificados como "ruído".
Resultado: Você pode identificar clusters que correspondem a diferentes bairros ou regiões com características de qualidade do ar distintas, e pode investigar os sensores classificados como ruído, que podem indicar falhas no sensor ou eventos de poluição localizados.
Conclusão
O aprendizado não supervisionado é uma ferramenta poderosa para descobrir insights em dados sem rótulos. O K-Means é excelente para clusters esféricos e de densidade similar, enquanto o DBSCAN se destaca na identificação de clusters de formas arbitrárias e na detecção de ruído. A escolha do algoritmo depende da natureza dos seus dados e dos objetivos do seu problema. Para dominá-los, a combinação de teoria sólida, prática com bibliotecas e a aplicação em problemas reais é a chave.

### Import das Bibliotecas

```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
```

# 1. Carregamento e Exploração dos Dados

```python
print("--- 1. Carregando e Explorando o Dataset Wine ---")
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['target'] = wine.target # Adicionamos o 'target' para comparação posterior, mas o K-Means não o usará

print("\nPrimeiras 5 linhas do DataFrame:")
print(df.head())
```

## Analisando os dados

```python
print(f"\nDimensões do DataFrame: {df.shape}")
print(f"Nomes das Características: {df.columns.tolist()}")
```

# 2. Pré-processamento: Padronização dos Dados

```python
# Excluímos a coluna 'target' da padronização, pois ela é o rótulo verdadeiro que queremos comparar
features = df.drop('target', axis=1)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
scaled_df = pd.DataFrame(scaled_features, columns=features.columns)

print("\nPrimeiras 5 linhas do DataFrame Padronizado:")
print(scaled_df.head())
```

# 3. Determinação do Número Ideal de Clusters (Método do Cotovelo)
# O método do cotovelo (Elbow Method) nos ajuda a encontrar um bom valor para 'k'
# Ele plota a soma dos quadrados das distâncias (SSE - Sum of Squared Errors)
# de cada ponto para seu centróide mais próximo para diferentes valores de 'k'.
# O "cotovelo" na curva indica um bom compromisso.

```python
print("\n--- 3. Determinando o Número Ideal de Clusters (Método do Cotovelo) ---")
sse = []
k_values = range(1, 11) # Testando de 1 a 10 clusters

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # n_init para mitigar problemas de inicialização
    kmeans.fit(scaled_df)
    sse.append(kmeans.inertia_) # 'inertia_' é a soma dos quadrados das distâncias
```

## Plotando a curva de elbow

```python
plt.figure(figsize=(10, 6))
plt.plot(k_values, sse, marker='o')
plt.title('Método do Cotovelo para K-Means no Dataset Wine')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Soma dos Quadrados das Distâncias (SSE)')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python
print("\nCom base no gráfico do cotovelo, podemos observar um 'cotovelo' aparente em k=3.")
print("Isso sugere que 3 clusters podem ser um número ideal para o nosso problema.")
print("Curiosamente, o dataset original tem 3 cultivares de vinho!")
```

# 4. Aplicação do K-Means com k=3

```python
print("\n--- 4. Aplicando K-Means com k=3 ---")
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(scaled_df)
df['cluster'] = clusters

print("\nContagem de pontos em cada cluster:")
print(df['cluster'].value_counts())
```

# 5. Análise dos Resultados: Visualização

### Principal Componente

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['cluster'] = df['cluster']
pca_df['target'] = df['target'] # Adiciona os rótulos originais para comparação
```

## Plotando as Análises

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x='PC1', y='PC2', hue='cluster', data=pca_df, palette='viridis', s=100, alpha=0.8, legend='full')
plt.title('Clusters K-Means (Visualização PCA)')
plt.xlabel('Primeira Componente Principal (PC1)')
plt.ylabel('Segunda Componente Principal (PC2)')
plt.grid(True)

plt.subplot(1, 2, 2)
sns.scatterplot(x='PC1', y='PC2', hue='target', data=pca_df, palette='tab10', s=100, alpha=0.8, legend='full')
plt.title('Rótulos Originais (Visualização PCA)')
plt.xlabel('Primeira Componente Principal (PC1)')
plt.ylabel('Segunda Componente Principal (PC2)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Interpretação dos resultados

```python
print("\n--- Comparação entre Clusters K-Means e Rótulos Originais ---")
# Criamos uma tabela cruzada para comparar os clusters formados pelo K-Means
# com os rótulos originais das cultivares de vinho.
# Isso nos ajuda a ver quão bem o K-Means "descobriu" os grupos inerentes aos dados.
comparison_table = pd.crosstab(df['target'], df['cluster'])
print("\nTabela de Contingência (Rótulos Originais vs. Clusters K-Means):")
print(comparison_table)

print("\nObservações:")
print("O K-Means conseguiu agrupar a maioria dos vinhos de forma bastante eficaz.")
print("Podemos ver que:")
print("- O cluster 0 do K-Means corresponde predominantemente à cultivar 1.")
print("- O cluster 1 do K-Means corresponde principalmente à cultivar 0.")
print("- O cluster 2 do K-Means corresponde em grande parte à cultivar 2.")
print("\nNo entanto, há algumas instâncias onde os vinhos foram agrupados de forma diferente (ex: alguns vinhos da cultivar 1 foram para o cluster 1, e vice-versa).")
print("Isso mostra que, embora o K-Means seja bom em encontrar padrões, a clusterização não supervisionada não é perfeita e a 'verdade' dos dados (rótulos originais) nem sempre é 100% recuperável apenas com base nas características.")
print("Apesar disso, o K-Means identificou com sucesso 3 grupos distintos, o que é um resultado excelente dada a natureza não supervisionada da tarefa!")
```

O dataset Wine é um conjunto de dados clássico no campo do machine learning, muito utilizado para demonstrar algoritmos de classificação e, como vimos, de clusterização. Ele contém informações de 178 amostras de vinhos, originárias de três diferentes cultivares (variedades de uva), que são o "rótulo" original que o algoritmo K-Means não tem acesso durante a clusterização.

Principais Características do Dataset Wine
Cada amostra de vinho no dataset é descrita por 13 características (atributos) químicas contínuas, que representam diferentes componentes do vinho. Entre elas, temos:

Essas 13 características são as "pistas" que o algoritmo utiliza para tentar encontrar padrões nos dados.



A decisão de agrupar em 3 clusters foi fortemente influenciada por dois fatores:

Estrutura Intrínseca dos Dados: Mesmo sem saber que existem 3 cultivares, a distribuição dos 178 vinhos no espaço de 13 dimensões apresentava três regiões de maior densidade ou concentração de pontos. O algoritmo, iterativamente, encontrou esses "aglomerados" naturais.

Método do Cotovelo (Elbow Method): Ao testar o K-Means com diferentes números de clusters (de 1 a 10), observamos que a Soma dos Quadrados das Distâncias (SSE) de cada ponto ao seu centróide caía drasticamente até k=3, e depois a queda se tornava menos acentuada. Esse ponto de inflexão na curva (o "cotovelo") é uma heurística comum que sugere o número ideal de clusters. A coincidência de o cotovelo ser em 3, exatamente o número de cultivares, é uma forte indicação de que o algoritmo estava capturando a estrutura subjacente dos dados.

## Síntese do Resultado Obtido pelo K-Means

A clusterização com K-Means no dataset Wine foi muito bem-sucedida. Apesar de ser um algoritmo de aprendizado não supervisionado (ou seja, ele não "viu" os rótulos das cultivares durante o treinamento), o K-Means conseguiu agrupar a grande maioria das amostras de vinho em três clusters distintos que correspondem de perto às três cultivares originais.

Outro cluster correspondeu, em sua maioria, aos vinhos da Cultivar 1.
O terceiro cluster agrupou a maioria dos vinhos da Cultivar 2.

```python
import pandas as pd
import numpy as np

# Definindo as características para a geração de dados
num_amostras = 500

# Faixa etária de 20 a 55 anos
idades = np.random.randint(20, 56, num_amostras)

# Renda em múltiplos de salários mínimos (considerando um salário mínimo de R$1.412,00 em 2024 para base)
# Distribuindo a renda para simular mais pessoas na faixa de 2 e 3 salários, e menos na de 4.
salario_minimo_base = 1412 # Valor de referência, pode ser ajustado
rendas_sm_multiplos = np.random.choice([2, 3, 4], size=num_amostras, p=[0.45, 0.40, 0.15])
rendas = rendas_sm_multiplos * salario_minimo_base + np.random.uniform(0, salario_minimo_base/2, num_amostras) # Adiciona variação dentro da faixa

# Moradia em bairro de classe média alta, periferia e suburbana
# Distribuindo para simular uma proporção realista (mais em subúrbio/periferia, menos em classe média alta)
bairros = np.random.choice(['Classe Média Alta', 'Periferia', 'Suburbana'], size=num_amostras, p=[0.25, 0.35, 0.40])

# Simulação de comportamento de compra (dias desde a última compra e valor total gasto)
# Para simular que alguns grupos compram mais que outros, vamos criar uma correlação sutil
# Por exemplo, pessoas mais jovens e de maior renda podem ter mais compras recentes e gastar mais

dias_desde_ultima_compra = np.zeros(num_amostras)
valor_total_gasto = np.zeros(num_amostras)

for i in range(num_amostras):
    # Simular que pessoas mais jovens e com maior renda tendem a ter menos dias desde a última compra
    # e maior valor gasto, mas com variabilidade.
    if idades[i] <= 35 and rendas_sm_multiplos[i] >= 3: # Jovens e com boa renda
        dias_desde_ultima_compra[i] = np.random.randint(1, 60) # Compras mais recentes
        valor_total_gasto[i] = np.random.uniform(500, 3000) # Gasto maior
    elif idades[i] > 45 and rendas_sm_multiplos[i] <= 2: # Mais velhos e com menor renda
        dias_desde_ultima_compra[i] = np.random.randint(90, 365) # Compras menos recentes
        valor_total_gasto[i] = np.random.uniform(100, 800) # Gasto menor
    else: # Outros perfis
        dias_desde_ultima_compra[i] = np.random.randint(30, 180)
        valor_total_gasto[i] = np.random.uniform(200, 1500)

# Arredondar valores de renda e gasto para duas casas decimais
rendas = np.round(rendas, 2)
valor_total_gasto = np.round(valor_total_gasto, 2)

# Criação do DataFrame
dados_consumidores = pd.DataFrame({
    'Idade': idades,
    'Renda_Salarios_Minimos': rendas_sm_multiplos,
    'Renda_Bruta_R$': rendas,
    'Bairro': bairros,
    'Dias_Desde_Ultima_Compra': dias_desde_ultima_compra.astype(int),
    'Valor_Total_Gasto_R$': valor_total_gasto
})

print("--- Amostra de Dados de Consumidores ---")
print("Primeiras 5 linhas do DataFrame:")
print(dados_consumidores.head())

print(f"\nDimensões do DataFrame: {dados_consumidores.shape}")

print("\nEstatísticas Descritivas:")
print(dados_consumidores.describe())

print("\nContagem de Moradia por Bairro:")
print(dados_consumidores['Bairro'].value_counts())

print("\nExemplo de uso: Qual a média de idade por faixa de renda em salários mínimos?")
print(dados_consumidores.groupby('Renda_Salarios_Minimos')['Idade'].mean().round(1))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Carregamento/Reutilização dos Dados ---
# Re-gerar a amostra de dados para garantir que tenhamos o DataFrame `dados_consumidores`
# sem depender da execução anterior do notebook/script.
num_amostras = 500
idades = np.random.randint(20, 56, num_amostras)
salario_minimo_base = 1412
rendas_sm_multiplos = np.random.choice([2, 3, 4], size=num_amostras, p=[0.45, 0.40, 0.15])
rendas = rendas_sm_multiplos * salario_minimo_base + np.random.uniform(0, salario_minimo_base/2, num_amostras)
bairros = np.random.choice(['Classe Média Alta', 'Periferia', 'Suburbana'], size=num_amostras, p=[0.25, 0.35, 0.40])

dias_desde_ultima_compra = np.zeros(num_amostras)
valor_total_gasto = np.zeros(num_amostras)

for i in range(num_amostras):
    if idades[i] <= 35 and rendas_sm_multiplos[i] >= 3:
        dias_desde_ultima_compra[i] = np.random.randint(1, 60)
        valor_total_gasto[i] = np.random.uniform(500, 3000)
    elif idades[i] > 45 and rendas_sm_multiplos[i] <= 2:
        dias_desde_ultima_compra[i] = np.random.randint(90, 365)
        valor_total_gasto[i] = np.random.uniform(100, 800)
    else:
        dias_desde_ultima_compra[i] = np.random.randint(30, 180)
        valor_total_gasto[i] = np.random.uniform(200, 1500)

rendas = np.round(rendas, 2)
valor_total_gasto = np.round(valor_total_gasto, 2)

dados_consumidores = pd.DataFrame({
    'Idade': idades,
    'Renda_Salarios_Minimos': rendas_sm_multiplos,
    'Renda_Bruta_R$': rendas,
    'Bairro': bairros,
    'Dias_Desde_Ultima_Compra': dias_desde_ultima_compra.astype(int),
    'Valor_Total_Gasto_R$': valor_total_gasto
})

print("--- Dados de Consumidores Carregados/Regerados ---")
print(dados_consumidores.head())

# --- 2. Pré-processamento: Padronização e One-Hot Encoding ---
# Definir as colunas numéricas e categóricas
numeric_features = ['Idade', 'Renda_Bruta_R$', 'Dias_Desde_Ultima_Compra', 'Valor_Total_Gasto_R$']
categorical_features = ['Renda_Salarios_Minimos', 'Bairro'] # Tratando 'Renda_Salarios_Minimos' como categórica ordenada

# Criar um pré-processador usando ColumnTransformer
# Scale as features numéricas e aplique One-Hot Encoding nas categóricas
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

# Aplicar o pré-processamento aos dados
scaled_data = preprocessor.fit_transform(dados_consumidores)

# Obter os nomes das colunas após o one-hot encoding
ohe_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)
all_feature_names = numeric_features + list(ohe_feature_names)

# Criar um DataFrame com os dados pré-processados
scaled_df = pd.DataFrame(scaled_data, columns=all_feature_names)

print("\n--- Dados Pré-processados (Padronizados e One-Hot Encoded) ---")
print(scaled_df.head())
print(f"Dimensões do DataFrame Pré-processado: {scaled_df.shape}")

# --- 3. Determinação do Número de Clusters (Método do Cotovelo) ---
print("\n--- 3. Determinando o Número Ideal de Clusters (Método do Cotovelo) ---")
sse = []
k_values = range(1, 11) # Testando de 1 a 10 clusters

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_df)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_values, sse, marker='o')
plt.title('Método do Cotovelo para K-Means - Dados de Consumidores')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Soma dos Quadrados das Distâncias (SSE)')
plt.xticks(k_values)
plt.grid(True)
plt.show()

print("\nCom base no gráfico do cotovelo, procure um ponto de inflexão significativo.")
print("Para estes dados simulados, um 'cotovelo' pode ser observado em k=3 ou k=4, sugerindo entre 3 a 4 clusters.")
# Vamos escolher k=4 para explorar um pouco mais, mas k=3 também seria uma escolha válida.
n_clusters_chosen = 4
print(f"Vamos prosseguir com {n_clusters_chosen} clusters.")

# --- 4. Aplicação do K-Means ---
print(f"\n--- 4. Aplicando K-Means com k={n_clusters_chosen} ---")
kmeans = KMeans(n_clusters=n_clusters_chosen, random_state=42, n_init=10)
clusters = kmeans.fit_predict(scaled_df)
dados_consumidores['Cluster'] = clusters

print("\nContagem de pontos em cada cluster:")
print(dados_consumidores['Cluster'].value_counts())

# --- 5. Análise e Visualização dos Clusters ---
print("\n--- 5. Análise e Visualização dos Clusters ---")

# Redução de dimensionalidade com PCA para visualização em 2D
pca = PCA(n_components=2, random_state=42)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['Cluster'] = dados_consumidores['Cluster'] # Adicionar os rótulos de cluster

# Visualização dos clusters usando Seaborn
plt.figure(figsize=(12, 8))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df,
                palette='viridis', s=100, alpha=0.8, legend='full')
plt.title(f'Clusters K-Means de Consumidores (Visualização PCA, k={n_clusters_chosen})')
plt.xlabel(f'Primeira Componente Principal (PC1) - Variância Explicada: {pca.explained_variance_ratio_[0]:.2f}')
plt.ylabel(f'Segunda Componente Principal (PC2) - Variância Explicada: {pca.explained_variance_ratio_[1]:.2f}')
plt.grid(True)
plt.show()

# Análise dos Centróides dos Clusters (para entender o perfil de cada grupo)
print("\n--- Análise dos Perfis dos Clusters ---")

# Para interpretar os centróides, é útil ver os valores médios das características originais (não padronizadas)
# para cada cluster.
cluster_profiles = dados_consumidores.groupby('Cluster').agg({
    'Idade': ['mean', 'min', 'max'],
    'Renda_Bruta_R$': ['mean', 'min', 'max'],
    'Dias_Desde_Ultima_Compra': ['mean', 'min', 'max'],
    'Valor_Total_Gasto_R$': ['mean', 'min', 'max'],
    'Renda_Salarios_Minimos': lambda x: x.mode()[0], # O salário mínimo mais frequente
    'Bairro': lambda x: x.mode()[0] # O bairro mais frequente
}).round(2)

# Renomear as colunas para facilitar a leitura
cluster_profiles.columns = ['_'.join(col).strip() for col in cluster_profiles.columns.values]
print(cluster_profiles)

print("\n--- Interpretação dos Clusters ---")
print("Analisando os perfis acima, podemos tentar descrever cada cluster:")
print("Cluster 0: Geralmente mais velhos, renda média-baixa, compras menos frequentes e menor gasto. Predominantemente Periferia.")
print("Cluster 1: Idade e renda intermediárias, padrão de compra e gasto moderado. Predominantemente Suburbana.")
print("Cluster 2: Clientes mais jovens, renda mais alta, compras mais recentes e maior gasto. Predominantemente Classe Média Alta.")
print("Cluster 3: Idade e renda mais alta que o cluster 1, com comportamento de compra e gasto ligeiramente melhor que o cluster 1. Predominantemente Suburbana ou Classe Média Alta.")
print("\nEsses são apenas exemplos de perfis, a interpretação exata dependerá dos dados simulados e do 'k' escolhido.")
print("Para um e-commerce, o Cluster 2 seria seu público 'gold' ou 'premium'.")
```
