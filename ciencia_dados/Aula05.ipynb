####Atividade 04

Leia a base de dados bancários (dataset) e realize as tratativas a seguir:

1.Remova a coluna ‘Unnamed: 0’.
2.Preencha os valores ‘NaN’ da coluna ‘saldo’ com a mediana.
3.Preencha os valores ‘NaN’ da coluna ‘genero’ com a moda.
4.Utilize a função .replace() para tratar os valores inválidos na coluna
‘genero’.
5.Utilize a função filter para remover as linhas que contém idades inválidas
(Maior que 120 e Menor que 0).

```python
bank = r'https://raw.githubusercontent.com/davi-lucciola/DatasetsEstudos/main/BankData/bank_data.csv'
df_bank = pd.read_csv(bank)
df_bank.head()
```

```python
df_bank.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# df_bank['saldo'].fillna(df_bank['saldo'].median(), inplace=True)
# df_bank.head()
```

```python
df_bank['saldo'].replace(0, df_bank['saldo'].median(), inplace=True)
df_bank.head()
```

```python
#moda = df_bank['saldo'].mode()

df_bank = df_bank[(df_bank['idade'] > 0) & (df_bank['idade'] <= 120)]
```

```python
df_bank.head()
```

```python
df_bank['genero'].replace({"Feminino": "F",
                      "Masculino": "M",
                      "Fem": "F"}, inplace=True)
```

```python
df_bank.head()
```

##Groupby()

```python
df_bank[["genero", "estado", "produtos"]].groupby(["genero", "estado"]).mean()
df_bank
```

```python
df_bank.head(2)
```

```python
#fillna() com o valor da moda captado corretamente (funcionou):
moda = df_bank['genero'].mode().iloc[0]
df_bank['genero'].fillna(moda, inplace=True)
df_bank.info()
```
